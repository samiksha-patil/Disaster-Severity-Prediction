```python
#importing data
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
```

# Finding Correlation between the data

```python
emdat = pd.read_excel('../Data/public_emdat_custom_request_2024-09-21_34981889-794d-4608-910c-8ce0702538b0.xlsx')

emdat_relevant = emdat[['Disaster Type', 'Location', 'Total Damage (\'000 US$)', 'Magnitude', 'Insured Damage (\'000 US$)',
                        'Total Deaths', 'AID Contribution (\'000 US$)', 'Total Affected', 'Reconstruction Costs (\'000 US$)']]


contingency_table = pd.crosstab(emdat_relevant['Disaster Type'], emdat_relevant['Location'])
contingency_table1 = pd.crosstab(emdat_relevant['Total Damage (\'000 US$)'], emdat_relevant['Magnitude'])
contingency_table2 = pd.crosstab(emdat_relevant['Total Affected'], emdat_relevant['Location'])
contingency_table3 = pd.crosstab(emdat_relevant['Total Deaths'], emdat_relevant['AID Contribution (\'000 US$)'])
contingency_table4 = pd.crosstab(emdat_relevant['Reconstruction Costs (\'000 US$)'], emdat_relevant['Total Deaths'])
contingency_table5 = pd.crosstab(emdat_relevant['Insured Damage (\'000 US$)'], emdat_relevant['Location'])

chi2, p, dof, expected = chi2_contingency(contingency_table)
chi2_1, p_1, dof_1, expected_1 = chi2_contingency(contingency_table1)
chi2_2, p_2, dof_2, expected_2 = chi2_contingency(contingency_table2)
chi2_3, p_3, dof_3, expected_3 = chi2_contingency(contingency_table3)
chi2_4, p_4, dof_4, expected_4 = chi2_contingency(contingency_table4)
chi2_5, p_5, dof_5, expected_5 = chi2_contingency(contingency_table5)

print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-Value:", p)

print("Chi-Square Statistic:", chi2_1)
print("Degrees of Freedom:", dof_1)
print("P_1-Value:", p_1)

print("Chi-Square Statistic:", chi2_2)
print("Degrees of Freedom:", dof_2)
print("P_2-Value:", p_2)

print("Chi-Square Statistic:", chi2_3)
print("Degrees of Freedom:", dof_3)
print("P_3-Value:", p_3)

print("Chi-Square Statistic:", chi2_4)
print("Degrees of Freedom:", dof_4)
print("P_4-Value:", p_4)

print("Chi-Square Statistic:", chi2_5)
print("Degrees of Freedom:", dof_5)
print("P_5-Value:", p_5)


if p < 0.05:
    print("There is a statistically significant association between Disaster Type and Location (p < 0.05).")
else:
    print("There is no statistically significant association between Disaster Type and Location (p >= 0.05).")
    
if p_1 < 0.05:
    print("There is a statistically significant association between Total Damage and Magnitude (p < 0.05).")
else:
    print("There is no statistically significant association between Total Damage and Magnitude (p >= 0.05).")

if p_2 < 0.05:
    print("There is a statistically significant association between Total Affected and Total Deaths (p < 0.05).")
else:
    print("There is no statistically significant association between Total Affected and Total Deaths (p >= 0.05).")
    
if p_3 < 0.05:
    print("There is a statistically significant association between Total Deaths and AID Contribution (p < 0.05).")
else:
    print("There is no statistically significant association between Total Deaths and AID Contribution (p >= 0.05).")
    
if p_4 < 0.05:
    print("There is a statistically significant association between Reconstruction Costs and Total Deaths (p < 0.05).")
else:
    print("There is no statistically significant association between Reconstruction Costs and Total Deaths (p >= 0.05).")
    
if p_5 < 0.05:
    print("There is a statistically significant association between Insured Damage and Total Damage (p < 0.05).")
else:
    print("There is no statistically significant association between Insured Damage and Total Damage (p >= 0.05).")
```

```python

emdat_relevant = emdat[['Disaster Type', 'Location']]

contingency_table = pd.crosstab(emdat_relevant['Disaster Type'], emdat_relevant['Location'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-Value:", p)


if p < 0.05:
    print("There is a statistically significant association between Disaster Type and Location (p < 0.05).")
else:
    print("There is no statistically significant association between Disaster Type and Location (p >= 0.05).")
```

# Visulization of correlated columns to understand the data

```python

emdat_relevant = emdat[['Disaster Type', 'Location']]

emdat_relevant['Region'] = emdat_relevant['Location'].str.split(',').str[-1].str.strip()

top_regions = emdat_relevant['Region'].value_counts().nlargest(10).index
top_region_disasters = emdat_relevant[emdat_relevant['Region'].isin(top_regions)]

region_disaster_counts = top_region_disasters.groupby(['Region', 'Disaster Type']).size().reset_index(name='Count')

plt.figure(figsize=(12, 8))
sns.barplot(data=region_disaster_counts, x='Count', y='Region', hue='Disaster Type', dodge=False)
plt.title("Most Frequent Disasters in Top 10 Regions")
plt.xlabel("Frequency of Disaster Occurrence")
plt.ylabel("Region")
plt.legend(title="Disaster Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Explanation of Code

	•	Region Extraction: We extract the last part of the Location column to approximate the region or country.
	•	Top 10 Regions: We identify the top 10 regions by disaster frequency.
	•	Frequency Table: We calculate the frequency of each disaster type within these top regions.
	•	Visualization: The bar plot shows the most frequent disaster types for each of these top regions, helping to visualize associations by location.

```python

from scipy.stats import skew

disaster_type_counts = emdat['Disaster Type'].value_counts().reset_index()
disaster_type_counts.columns = ['Disaster Type', 'Count']

plt.figure(figsize=(12, 6))
sns.barplot(data=disaster_type_counts, x='Count', y='Disaster Type', palette='viridis')
plt.title("Distribution of Data by Disaster Type")
plt.xlabel("Number of Records")
plt.ylabel("Disaster Type")
plt.tight_layout()
plt.show()

skewness_value = skew(disaster_type_counts['Count'])
print(f"Skewness of the distribution: {skewness_value:.2f}")

if skewness_value > 1:
    print("The data is highly positively skewed, indicating a strong bias toward certain disaster types.")
elif 0.5 < skewness_value <= 1:
    print("The data is moderately positively skewed, indicating some bias toward certain disaster types.")
elif -1 < skewness_value <= -0.5:
    print("The data is moderately negatively skewed, indicating some bias against certain disaster types.")
elif skewness_value <= -1:
    print("The data is highly negatively skewed, indicating a strong bias against certain disaster types.")
else:
    print("The data is fairly symmetric, with no significant skew.")
```

## visulaization of earhquake data

```python
data = pd.read_csv("../Data/earthquakes-2024-11-12_20-00-06_-0700.tsv", sep="\t")

data['Datetime'] = pd.to_datetime(data['Year'].astype(str) + '-' + data['Mo'].astype(str).str.zfill(2) + '-' + data['Dy'].astype(str).str.zfill(2) + ' ' + data['Hr'].astype(str).str.zfill(2) + ':' + data['Mn'].astype(str).str.zfill(2) + ':' + data['Sec'].astype(str).str.zfill(2), errors='coerce')

data.replace({" ": None, "": None}, inplace=True)  
data = data.apply(pd.to_numeric, errors="ignore")  

plt.figure(figsize=(10, 6))
sns.histplot(data['Mag'], bins=30, kde=True, color='blue')
plt.title("Distribution of Earthquake Magnitudes")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
# Geospatial Distribution of Earthquakes
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', hue='Mag', size='Mag', sizes=(50, 200), data=data, palette='coolwarm', alpha=0.6)
plt.title("Geospatial Distribution of Earthquakes by Magnitude")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Magnitude", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Correlation Heatmap for Numeric Features
numeric_cols = ['Focal Depth (km)', 'Mag', 'MMI Int', 'Deaths', 'Injuries', 'Damage ($Mil)', 'Houses Destroyed', 'Total Deaths', 'Total Damage ($Mil)']
correlation_matrix = data[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Earthquake Magnitude vs. Damage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mag', y='Total Damage ($Mil)', hue='Mag', size='Mag', sizes=(20, 200), data=data, palette='viridis', alpha=0.7)
plt.title("Magnitude vs. Total Damage")
plt.xlabel("Magnitude")
plt.ylabel("Total Damage (in $ Million)")
plt.show()

# Top Locations by Earthquake Count
top_locations = data['Location Name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_locations.plot(kind='bar', color='purple')
plt.title("Top 10 Locations by Earthquake Count")
plt.xlabel("Location")
plt.ylabel("Number of Earthquakes")
plt.xticks(rotation=45, ha='right')
plt.show()

# Deaths and Injuries by Magnitude
plt.figure(figsize=(12, 6))
sns.boxplot(x='Mag', y='Deaths', data=data, palette='Reds')
plt.title("Deaths Distribution by Magnitude")
plt.xlabel("Magnitude")
plt.ylabel("Number of Deaths")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Mag', y='Injuries', data=data, palette='Blues')
plt.title("Injuries Distribution by Magnitude")
plt.xlabel("Magnitude")
plt.ylabel("Number of Injuries")
plt.show()

# Earthquake Intensity (MMI) over Time
plt.figure(figsize=(14, 6))
sns.lineplot(x='Datetime', y='MMI Int', data=data, marker='o', color='green')
plt.title("Earthquake Intensity (MMI) Over Time")
plt.xlabel("Time")
plt.ylabel("MMI Int")
plt.show()

# Damage vs. Depth
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Focal Depth (km)', y='Total Damage ($Mil)', hue='Mag', size='Mag', sizes=(50, 200), data=data, palette='plasma', alpha=0.7)
plt.title("Focal Depth vs. Total Damage")
plt.xlabel("Focal Depth (km)")
plt.ylabel("Total Damage (in $ Million)")
plt.show()
```

```python

tsunami_data = pd.read_csv("../Data/tsunamis-2024-11-12_19-46-31_-0700.tsv", sep="\t")

tsunami_data['Year'] = pd.to_numeric(tsunami_data['Year'], errors='coerce')
tsunami_data['Total Damage ($Mil)'] = pd.to_numeric(tsunami_data['Total Damage ($Mil)'], errors='coerce')
tsunami_data['Total Deaths'] = pd.to_numeric(tsunami_data['Total Deaths'], errors='coerce')
tsunami_data['Total Injuries'] = pd.to_numeric(tsunami_data['Total Injuries'], errors='coerce')
tsunami_data['Earthquake Magnitude'] = pd.to_numeric(tsunami_data['Earthquake Magnitude'], errors='coerce')
tsunami_data['Maximum Water Height (m)'] = pd.to_numeric(tsunami_data['Maximum Water Height (m)'], errors='coerce')

tsunami_data = tsunami_data.dropna(subset=['Year'])

plt.figure(figsize=(14, 8))
for metric, label, color in zip(
    ['Total Deaths', 'Total Injuries', 'Total Damage ($Mil)'], 
    ['Total Deaths', 'Total Injuries', 'Total Damage (Million $)'], 
    ['red', 'blue', 'green']
):
    sns.lineplot(
        data=tsunami_data.groupby('Year', as_index=False)[metric].sum(),
        x='Year', y=metric, label=label, color=color
    )

plt.title('Impact of Tsunamis Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Count / Damage ($Mil)', fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

top_countries = tsunami_data.groupby('Country', as_index=False)['Total Deaths'].sum().sort_values('Total Deaths', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x='Total Deaths', y='Country', palette='viridis')
plt.title('Top 10 Countries by Total Deaths', fontsize=16)
plt.xlabel('Total Deaths', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.tight_layout()
plt.show()


```

```python
columns_to_numeric = [
    'Year', 'Total Damage ($Mil)', 'Total Deaths', 'Total Injuries', 
    'Earthquake Magnitude', 'Maximum Water Height (m)'
]
for col in columns_to_numeric:
    tsunami_data[col] = pd.to_numeric(tsunami_data[col], errors='coerce')

tsunami_data = tsunami_data.dropna(subset=['Year'])

tsunami_data[['Total Deaths', 'Total Injuries', 'Total Damage ($Mil)']] = tsunami_data[
    ['Total Deaths', 'Total Injuries', 'Total Damage ($Mil)']
].fillna(0)

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=tsunami_data, 
    x='Earthquake Magnitude', 
    y='Total Damage ($Mil)', 
    size='Total Deaths', 
    hue='Maximum Water Height (m)', 
    palette='coolwarm', 
    sizes=(20, 200), alpha=0.6, edgecolor="k"
)
plt.title('Correlation Between Earthquake Magnitude and Total Damage', fontsize=16)
plt.xlabel('Earthquake Magnitude', fontsize=14)
plt.ylabel('Total Damage ($Mil)', fontsize=14)
plt.yscale('log')  # Log scale to handle large range in damage values
plt.legend(title='Max Water Height (m)', loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

## Relationship Between Maximum Water Height and Total Deaths
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=tsunami_data, 
    x='Maximum Water Height (m)', 
    y='Total Deaths', 
    hue='Country', 
    palette='tab20', 
    alpha=0.8, edgecolor="k"
)
plt.title('Relationship Between Maximum Water Height and Total Deaths', fontsize=16)
plt.xlabel('Maximum Water Height (m)', fontsize=14)
plt.ylabel('Total Deaths', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()

## Tsunami Causes and Their Impact
cause_impact = tsunami_data.groupby('Tsunami Cause Code', as_index=False)['Total Deaths'].sum().sort_values(
    'Total Deaths', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=cause_impact, x='Total Deaths', y='Tsunami Cause Code', palette='magma')
plt.title('Tsunami Causes and Their Impact (Total Deaths)', fontsize=16)
plt.xlabel('Total Deaths', fontsize=14)
plt.ylabel('Cause Code', fontsize=14)
plt.tight_layout()
plt.show()
```
